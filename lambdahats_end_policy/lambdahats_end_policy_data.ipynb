```python
import lambdahat_helpers as lah
import time
import os
import torch
import numpy as np
import pickle
import argparse
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
file_0_2000 = "checkpoint_stats/end_policy/0-2000/losses.pickle"
with open(file_0_2000, "rb") as f:
    losses_0_2000 = pickle.load(f)
x_vals_0_2000 = np.arange(0, len(losses_0_2000), 1)

file_2000_4000 = "checkpoint_stats/end_policy/2000-4000/losses.pickle"
with open(file_2000_4000, "rb") as f:
    losses_2000_4000 = pickle.load(f)
x_vals_2000_4000 = np.arange(2000, 2000+len(losses_2000_4000), 1)

file_4000_6000 = "checkpoint_stats/end_policy/4000-6000/losses.pickle"
with open(file_4000_6000, "rb") as f:
    losses_4000_6000 = pickle.load(f)
x_vals_4000_6000 = np.arange(4000, 4000+len(losses_4000_6000), 1)

file_6000_8000 = "checkpoint_stats/end_policy/6000-8000/losses.pickle"
with open(file_6000_8000, "rb") as f:
    losses_6000_8000 = pickle.load(f)
x_vals_6000_8000 = np.arange(6000, 6000+len(losses_6000_8000), 1)

plt.plot(x_vals_0_2000, losses_0_2000, label="0-2000")
plt.plot(x_vals_2000_4000, losses_2000_4000, label="2000-4000")
plt.plot(x_vals_4000_6000, losses_4000_6000, label="4000-6000")
plt.plot(x_vals_6000_8000, losses_6000_8000, label="6000-8000")
plt.legend()
plt.show()
```

The model gets very bad at first, then very good again. Probably the reason is due to the effective horizon, where the optimal action given the optimal policy is equal to the optimal action given a random policy.
